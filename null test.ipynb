```python
import matplotlib.pyplot as plt
import numpy as np
import pysm
from pysm.nominal import models
from pysm.common import convert_units
import healpy as hp
%config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

```python
c1_config = models("c1", 1024)
sky_config = {'cmb' : c1_config}
sky = pysm.Sky(sky_config)
cmb = sky.signal()(30)
cmb = cmb*convert_units("uK_RJ", "uK_CMB", 30)
cl = hp.anafast(cmb[0])
cmb_map = hp.synfast(cl, 1024)
cl_real = hp.anafast(cmb_map)
```

```python
ell = np.arange(len(cl))
plt.plot(ell, ell*(ell+1)*cl/2/np.pi)
plt.plot(ell, ell*(ell+1)*cl_real/2/np.pi)
plt.yscale('log')
```
