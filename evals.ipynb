```python
#import sys 
#sys.path.append('/home/yao/Desktop/EM-smica/smica-codes/programs') 
#from smica_simulation_new import power_spectrum
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt

from NPTFit import create_mask as cm
mask = cm.make_mask_total(512,band_mask=True, band_mask_range = 10)
z = np.ones(len(mask))
fr = 1-np.dot(mask,z)/len(mask)

def bin_l(cl):
    bin_averages = []
    for q in range(Q):
        bin_averages.append(sum(cl[q*L/Q:((q+1)*L/Q)]/(L/Q)))
    return bin_averages
        
L = 1500; Q=50;Nf=7; E_cut=0.5; D_B = [];  SamNum = 1;  Evals = np.ones((Q,Nf));Delta = 0.00434117706763*20

for n in range(SamNum):
    D_B_n = []; 
    D = np.load('/home/yao/Desktop/EM-smica/ABS/simulation/test/total_power_synch*9.npy')
    noise_ps = np.load('/home/yao/Desktop/EM-smica/ABS/simulation/noise_power_spectrum.npy')
    CMB = np.load('/home/yao/Desktop/EM-smica/ABS/simulation/cmb_power_spectrum_masked.npy')
    
    D = bin_l(D); CMB = bin_l(CMB); noise_ps = bin_l(noise_ps*fr)  #*fr
    noise = np.zeros_like(noise_ps);f = []
    for i in range(Q):
        f_q = np.ones(Nf)
        for j in range(Nf):
            noise[i][j,j] = noise_ps[i][j,j]*np.sqrt(2.0/((2*i+1)*fr)*L/Q) # *fr
            f_q[j] = f_q[j]/np.sqrt(noise[i][j,j])
        f.append(f_q)
          
    for l in range(Q):
        D[l] = D[l] - noise_ps[l]
        for i in range(Nf): 
            for j in range(Nf):
                D[l][i,j] = D[l][i,j]/np.sqrt(noise[l][i,i]*noise[l][j,j]) + Delta*f[l][i]*f[l][j] 
                
    for l in range(Q): 
        e_vals,E = LA.eig(D[l])
        if n == 0:
            Evals[l,:] = e_vals
        for i in range(Nf):
            E[:,i]=E[:,i]/LA.norm(E[:,i])**2  
        
        D_B_l = 0
        for i in range(Nf):
            if e_vals[i]>=E_cut:
                G_i = np.dot(f[l],E[:,i])
                D_B_l = D_B_l + (G_i**2/e_vals[i]) 
        D_B_l = 1/ D_B_l - Delta
        D_B_n.append(D_B_l)
    D_B.append(D_B_n)
#np.save('/home/yao/Desktop/EM-smica/ABS/results/D_B_1.npy',D_B)

# plt.figure(figsize=(20,15))
for n in range(SamNum):           
    ell = np.ones(Q)
    for q in range(Q):
        ell[q] = (2*q+1)*L/Q/2 
#     plt.plot(ell,ell*(ell+1)*D_B[n]/(2*np.pi),'g-^') #,label = 'our CMB'

# plt.plot(ell,ell*(ell+1)*CMB[0:Q]/(2*np.pi),'r-x',label = 'real CMB')
# fig2 = plt.figure(2)
# plt.figure(figsize=(20,15))
# Error = [];s=1
# for i in range(s,Q):
#     error = (D_B[0][i]-CMB[i])/CMB[i]*100
#     Error.append(error)
# plt.plot(ell[s:Q],Error)
#np.save('/home/yao/Desktop/EM-smica/ABS/results/Errors_1',Error)
fig3 = plt.figure(3)
# plt.figure(figsize=(20,15))
for i in range(Nf):
    evals = Evals.T
    x = np.arange(len(evals[i]))
    plt.scatter(ell,np.log10(evals[i][:]))
    plt.axhline(0,color = 'k')
    
# fig4 = plt.figure(4)
# plt.figure(figsize=(20,15))
# z = []
# for i in range(50):
#     z.append(Evals[i].min())

# plt.plot(ell,z)
# plt.title('The minimum eigenvalues of each l_bin')

# #cl = np.ones(1500)*0.028**2
# #zll = np.ones(1500)*0.015**2
# #ell = np.arange(1500)
# #plt.plot(ell,ell*(ell+1)*cl/2/np.pi,label = '95')
# #plt.plot(ell,ell*(ell+1)*zll/2/np.pi,label = '150')
# plt.legend()
plt.show()
```
